```python
from tensorflow.keras.datasets import mnist
```

```python
(x_train , y_train) , (x_test, y_test) = mnist.load_data()
```

```python
import matplotlib.pyplot as plt
```

```python
plt.imshow(x_train[0], cmap="gray")
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
x_train, x_test = x_train/255.0, x_test / 255.0
x_train = x_train.reshape((x_train.shape[0], 28*28))
x_test = x_test.reshape((x_test.shape[0], 28*28))
```

```python
logreg = LogisticRegression( max_iter = 100)
from sklearn.metrics import accuracy_score
```

```python
logreg.fit(x_train, y_train)
```

```python
y_pred_logreg = logreg.predict(x_test)
accuracy_model = accuracy_score(y_test ,y_pred_logreg)
print("Logistic Regression Accuracy:" , accuracy_model)
```

```python
from sklearn.metrics import confusion_matrix
import seaborn as sns
```

```python
cn_svc = confusion_matrix(y_test,y_pred_logreg)
```

```python
sns.heatmap(cn_svc , annot=True , cmap='Blues' , xticklabels = 'true' , yticklabels = 'predicted')
plt.xlabel('Predicted')
plt.ylabel('True')
plt.title('Confusion Matrix')
```
